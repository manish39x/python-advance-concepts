### Set

```python
s = {1,2,3,4,3}
s.add(5),
s.discard(2)
# s.remove(34) # KeyError if not found
```

```python
a = {1,2,3}; b = {3,4,5,6}
print(a | b) # union
print(a & b) # intersection
print(a - b) # diff 
print(a ^ b) # symetric diff
print((a&b) <= b) # subset
```
