### Decorators

```python
from abc import ABC, abstractmethod

class Shape(ABC):
  @abstractmethod
  def area(self) -> float: ...

  @abstractmethod
  def perimeter(self) -> float: ...

class Rectangle(Shape):
  def __init__(self, w, h): self.w , self.h = w, h
  def area(self): return self.w * self.h
  def perimeter(self): return 2*(self.w+self.h)
```

```python
import functools
import time

def timer(func):
  @functools.wraps(func)
  def wrapper(*args, **kwargs):
    try:
      start_time = time.perf_counter()
      result = func(*args, **kwargs)
    finally:
      end_time = time.perf_counter()
      print(f"{func.__name__}: {end_time - start_time:.4f}s")
    return result
  return wrapper
```

```python
@timer
def compute_heavy_geometry():
  time.sleep(1.5)
  rectangles = [Rectangle(i, i**2) for i in range(10000)]
  total_area = sum(r.area() for r in rectangles)
  # print(5/0)
  return total_area

compute_heavy_geometry()
```

```python
# Decorator with arguments
def repeat(n):
  def decorator(func):
    @functools.wraps(func)
    def wrapper(*args, **kwargs):
      for _ in range(n): func(*args, **kwargs)
    return wrapper
  return decorator


@repeat(4)
def greet(name): print(f"Hi {name}")
```

```python
greet("manish")
```

```python
from contextlib import contextmanager

@contextmanager
def managed_file(path: str, mode: str):
  file = open(path, mode)
  try:
    yield file
  finally:
    file.close
```

```python
def make_multiplier(n):
    def multiply(x):
        return x * n   # captures n
    return multiply

triple = make_multiplier(3)
triple(7)  # → 21
```

```python
import functools
mul = lambda x, y: x ** y
cube = functools.partial(mul, y=3)

@functools.lru_cache(maxsize=128)
def fib(n):
    if n < 2: return n
    return fib(n-1) + fib(n-2)
```

```python
cube(5)
```

```python
from typing import TypeVar, Generic, Optional, Union, Callable

T = TypeVar('T')

class Stack(Generic[T]):
  def __init__(self) -> None:
    self._items: list[T] = []
  
  def push(self, item: T) -> None:
    self._items.append(item)
  
  def pop(self) -> Optional[T]:
    return self._items.pop() if self._items else None

def apply(fn: Callable[[int], int], values: list[int]) -> list[int]:
  return [fn(v) for v in values]
```

```python
num_stack = Stack[int]()
num_stack.push(3)
num_stack.push(4)
num_stack.push(5)

num_stack.pop()
```

```python
class Meta(type):
    def __new__(mcs, name, bases, namespace, **kwargs):
        namespace['greet'] = lambda self: f"I am {self.__class__.__name__}"
        return super().__new__(mcs, name, bases, namespace)

class Base(metaclass=Meta):
    def __init_subclass__(cls, prefix="", **kw):
        super().__init_subclass__(**kw)
        cls.greet = lambda self: f"{prefix}{cls.__name__}"

class Child(Base, prefix=">> "):
    pass

print(Child().greet())
print(Base().greet())
```
