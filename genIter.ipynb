### Iterator protocol

```python
class Countdown:
  def __init__(self, n): self.n = n
  def __iter__(self): return self
  def __next__(self):
    if self.n <= 0: raise StopIteration
    self.n -= 1; return self.n+1

for x in Countdown(10): print(x)
```

```python
## Generator Function
import math
def is_prime(n):
  if n <= 1: return False
  if n <= 3: return True
  for i in range(2, int(math.sqrt(n))+1):
    if n % i == 0:
      return False
  return True



def infinite_prime():
  current_prime = 2
  while True:
    if is_prime(current_prime):
      yield current_prime
    current_prime += 1

prime_gen = infinite_prime()
for i in range(10):
  print(next(prime_gen))
```

```python
def chain(*iterables):
  for it in iterables:
    try:
      yield from it
    except TypeError:
      yield it 

for item in chain([1,2,3], "ABC", 3):
  print(item, end=" ")
```

```python
def accumulator():
  total = 0
  while True:
    value = yield total
    if value is not None:
      total += value
  
acc = accumulator()
next(acc)
print(acc.send(2))
print(acc.send(2))
print(next(acc))
print(next(acc))
print(acc.send(5))
```

```python
import itertools as it

chain_it = it.chain([1,2], [3,4])
for item in chain_it:
  print(item)
```

```python
com = it.combinations('ABC', 2)       # AB AC BC
per = it.permutations('ABC', 2)       # all ordered pairs
```

```python
for item in com:
  print(item, end=" ")
print("")
for item in per:
  print(item, end=" ")
print("")
```
